## Creating Dummy variables

```python
##UCV = ['CarName', 'fueltype','aspiration','carbody','drivewheel','enginelocation','enginetype','fuelsystem']
##OCV = ['symboling','doornumber','cylindernumber',]

CarNameD = pd.get_dummies(CarPrice['CarName'], drop_first = True)
print(CarNameD)
CarPrice = pd.concat([CarPrice,CarNameD], axis = 1)
CarPrice.drop(['CarName'], axis = 1, inplace = True)
CarPrice.head()
```

```python
fueltypeD = pd.get_dummies(CarPrice['fueltype'], drop_first = True)
print(fueltypeD.head())
CarPrice = pd.concat([CarPrice,fueltypeD], axis = 1)
CarPrice.drop(['fueltype'], axis = 1, inplace = True)
CarPrice.head()
```

```python

aspirationD = pd.get_dummies(CarPrice['aspiration'], drop_first = True)
print(aspirationD.head())
CarPrice = pd.concat([CarPrice,aspirationD], axis = 1)
CarPrice.drop(['aspiration'], axis = 1, inplace = True)
CarPrice.head()
```

```python
carbodyD = pd.get_dummies(CarPrice['carbody'], drop_first = True)
print(carbodyD.head())
CarPrice = pd.concat([CarPrice,carbodyD], axis = 1)
CarPrice.drop(['carbody'], axis = 1, inplace = True)
CarPrice.head()
```

```python
drivewheelD = pd.get_dummies(CarPrice['drivewheel'], drop_first = True)
print(drivewheelD.head())
CarPrice = pd.concat([CarPrice,drivewheelD], axis = 1)
CarPrice.drop(['drivewheel'], axis = 1, inplace = True)
CarPrice.head()
```

```python

enginelocationD = pd.get_dummies(CarPrice['enginelocation'], drop_first = True)
print(enginelocationD.head())
CarPrice = pd.concat([CarPrice,enginelocationD], axis = 1)
CarPrice.drop(['enginelocation'], axis = 1, inplace = True)
CarPrice.head()
```

```python

enginetypeD = pd.get_dummies(CarPrice['enginetype'], drop_first = True)
print(enginetypeD.head())
CarPrice = pd.concat([CarPrice,enginetypeD], axis = 1)
CarPrice.drop(['enginetype'], axis = 1, inplace = True)
CarPrice.head()
```

```python
fuelsystemD = pd.get_dummies(CarPrice['fuelsystem'], drop_first = True)
print(fuelsystemD.head())
CarPrice = pd.concat([CarPrice,fuelsystemD], axis = 1)
CarPrice.drop(['fuelsystem'], axis = 1, inplace = True)
CarPrice.head()
```

```python
symbolingD = pd.get_dummies(CarPrice['symboling'], drop_first = True)
print(symbolingD.head())
```

```python
symbolingD.rename(columns=lambda x: str(x) + 'Sym', inplace=True)
symbolingD.columns
```

```python
CarPrice = pd.concat([CarPrice,symbolingD], axis = 1)
CarPrice.drop(['symboling'], axis = 1, inplace = True)
CarPrice.head()
```

```python
doornumberD = pd.get_dummies(CarPrice['doornumber'], drop_first = True)
doornumberD.head()
```

```python
doornumberD.rename(columns=lambda x: str(x) + 'Door', inplace=True)
doornumberD.columns
```

```python
CarPrice = pd.concat([CarPrice,doornumberD], axis = 1)
CarPrice.drop(['doornumber'], axis = 1, inplace = True)
CarPrice.head()
```

```python
cylindernumberD = pd.get_dummies(CarPrice['cylindernumber'], drop_first = True)
cylindernumberD.head()
```

```python
cylindernumberD.rename(columns=lambda x: str(x) + 'Cylinder', inplace=True)
cylindernumberD.columns
```

```python
CarPrice = pd.concat([CarPrice,cylindernumberD], axis = 1)
CarPrice.drop(['cylindernumber'], axis = 1, inplace = True)
CarPrice.head()
```

```python
from sklearn.model_selection import train_test_split
np.random.seed(0)
CarPrice_train, CarPrice_test = train_test_split(CarPrice, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
CarPrice_train[NV] = scaler.fit_transform(CarPrice_train[NV])
```

```python
y_train = CarPrice_train.pop('price')
x_train = CarPrice_train
```

```python
# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE with the output number of the variable equal to 10
lm = LinearRegression()
lm.fit(x_train, y_train)

rfe = RFE(lm, 10)             # running RFE
rfe = rfe.fit(x_train, y_train)
```

```python
list(zip(x_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = x_train.columns[rfe.support_]
col
```

### Running Regression Model

```python
x_train_rfe = x_train[col]
```

```python
import statsmodels.api as sm  
x_train_rfe = sm.add_constant(x_train_rfe)
```

```python
lm = sm.OLS(y_train,x_train_rfe).fit()
print(lm.summary())
```

### Removing variable with p value > 5

```python
x_train_new = x_train_rfe.drop(["twelveCylinder"], axis = 1)
```

```python
import statsmodels.api as sm  
x_train_lm = sm.add_constant(x_train_new)
lm = sm.OLS(y_train,x_train_lm).fit()   # Running the linear model again
print(lm.summary())
```

```python
x_train_new = x_train_new.drop(['const'], axis=1)
```

```python
# Calculate the VIFs for the new model
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = x_train_new
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```
