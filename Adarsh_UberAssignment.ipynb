## Working on Request Time 

```python
# Covert the request and drop time column to date format
UberData['Request timestamp'] = pd.to_datetime(UberData['Request timestamp'])
UberData['Drop timestamp'] = pd.to_datetime(UberData['Drop timestamp'])
UberData.info()
```

```python
# Function to calculate time of the day based on hour
def TimeOfDay(x) :
    if x in ([2,3,4,5]) :
        return 'Early Morning'
    elif x in ([6,7,8,9,10,11]) :
        return 'Morning'
    elif x in ([12,13,14,15,16]) :
        return 'Afternoon'
    elif x in ([17,18,19,20,]) :
        return 'Evening'
    else :
        return 'Night'
```

```python
UberData['Booking Month'] = UberData['Request timestamp'].dt.month # extracting month from request time
UberData['Booking Day'] = UberData['Request timestamp'].dt.dayofweek # extracting Day of week from request time
UberData['Booking Hour'] = UberData['Request timestamp'].dt.hour # extracting Hour of the day from request time
UberData['Booking Time'] = UberData['Booking Hour'].apply(lambda x : TimeOfDay(x)) # extracting Time of the day from hour

UberData.info()

#derived dataframes
UberData_CompletedTrips = UberData[UberData['Trip Completed'] == 1] # booking that are completed
UberData_FailedTrips = UberData[UberData['Trip Completed'] == 0]  # bookings that are failed
UberData_SuppliedTrips = UberData[UberData['Status'] != 'No Cars Available'] # bookings where cars are availabile
UberData_UnSuppliedTrips = UberData[UberData['Status'] == 'No Cars Available'] # bookings where cars are unavailable
```

## Demand Vs Supply

```python

Demand = [UberData['Request id'].count()] #height of the bar
Supply = [UberData_SuppliedTrips['Request id'].count()]#height of the bar

# positions on x axis
r1 = np.arange(len(Demand))
r2 = [x + 0.25 for x in r1]

# Plotting
plt.bar(r1,Demand,width = 0.25, label = 'Demand')
plt.bar(r2,Supply,width = 0.25, label = 'Supply')

plt.legend()
plt.show()

print('percentage of supply :' )
print((UberData_SuppliedTrips['Request id'].count()/UberData['Request id'].count())*100)
```

### Clearly Uber is not able to meet with the Supply. It is only able to supply 60 % of the total  demand.

### Demand Vs Supply based on time of day

```python
temp1 = UberData.groupby('Booking Time')
UberData_TimeGrouped = pd.DataFrame(temp1['Request id'].count())
print(UberData_TimeGrouped)

temp2 = UberData_CompletedTrips.groupby('Booking Time')
UberDataCT_TimeGrouped = pd.DataFrame(temp2['Request id'].count())
print(UberDataCT_TimeGrouped)

temp3 = UberData_SuppliedTrips.groupby('Booking Time')
UberDataST_TimeGrouped = pd.DataFrame(temp2['Request id'].count())
print(UberDataST_TimeGrouped)
```

```python
Demand = list(UberData_TimeGrouped['Request id'])
Supply = list(UberDataST_TimeGrouped['Request id'])

r1 = np.arange(len(Demand))
r2 = [x + 0.25 for x in r1]


plt.bar(r1,Demand,width = 0.25, label = 'Demand')
plt.bar(r2,Supply,width = 0.25, label = 'Supply')

plt.xlabel('group', fontweight='bold')
plt.xticks([r + 0.25 for r in range(len(Demand))], list(UberData_TimeGrouped.index))

plt.legend()
plt.show()
```

### As you can see, Demand is highest in mornings and evening. Also, Demand Supply gap is highest in the morning amd evening
