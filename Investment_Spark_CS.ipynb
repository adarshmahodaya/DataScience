```python
#split category list in master data frame to get primary sector
master_frame['Primary_Sector']=master_frame['category_list'].str.split('|').str[0]
#check the data if data is split accordingly
master_frame[['permalink','category_list','Primary_Sector']].head()

master_frame.rename(columns={"key_0":"key_1"},inplace=True)
master_frame.head()
```

```python
#load Sector mapping data
file3 = os.path.join(ROOTDIR, 'mapping.csv') 
mapping=pd.read_csv(file3,sep = ",", encoding = "ISO-8859-1")
mapping.shape
#(688, 10)
```

```python
#create row function to get Main Sector
def label_main_sect(row):
    if row['Automotive & Sports'] == 1 :
      return 'Automotive & Sports'
    if row['Blanks'] == 1 :
      return 'Blanks'
    if row['Cleantech / Semiconductors'] == 1 :
      return 'Cleantech / Semiconductors'
    if row['Entertainment'] == 1:
      return 'Entertainment'
    if row['Health']  == 1:
      return 'Health'
    if row['Manufacturing'] == 1:
      return 'Manufacturing'
    if row['News, Search and Messaging'] == 1:
      return 'News, Search and Messaging'
    if row['Social, Finance, Analytics, Advertising'] == 1:
      return 'Social, Finance, Analytics, Advertising'
    return 'Other'
```

```python
mapping['Main_Sector']=mapping.apply (lambda row: label_main_sect (row),axis=1)
mapping.head()
```

```python
#Create New data frame merging master df and mapping to get main sector
master_mapping_frame=pd.merge(master_frame, mapping[['category_list','Main_Sector']],  left_on = master_frame['Primary_Sector'].str.upper(), right_on = mapping['category_list'].str.upper(),how='left').reset_index(drop=True)

#find the observations of master_mapping_frame
master_mapping_frame.shape
#(89108, 21)
```

```python
#DF for USA and FT = Venture for Sector-wise analysis
D1=master_mapping_frame.loc[(master_mapping_frame.country_code == 'USA') & (master_mapping_frame.funding_round_type=='venture') 
                            & (master_mapping_frame.raised_amount_usd <= 15000000) & (master_mapping_frame.raised_amount_usd >= 5000000)]
D1.count()
#12150
D1['raised_amount_usd'].agg(['sum']).apply(lambda x: format(x, 'f'))
#108531347515.000000
D1.groupby(['Main_Sector']).count().sort_values(by='key_0',ascending=False)['raised_amount_usd']
#other
D1[D1['Main_Sector']=='Other'].groupby(['name']).count().sort_values(by='key_0',ascending=False)['raised_amount_usd'].head(5)
D1[D1['Main_Sector']=='Social, Finance, Analytics, Advertising'].groupby(['name']).count().sort_values(by='key_0',ascending=False)['raised_amount_usd'].head(5)
```
